```python
#공식 github 코드
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
from matplotlib import pyplot as plt

url = 'https://finance.naver.com/item/sise_day.nhn?code=068270&page=1'
with urlopen(url) as doc:
    html = BeautifulSoup(doc, 'lxml') 
    pgrr = html.find('td', class_='pgRR')
    s = str(pgrr.a['href']).split('=')
    last_page = s[-1]  

df = pd.DataFrame()
sise_url = 'https://finance.naver.com/item/sise_day.nhn?code=068270'  
for page in range(1, int(last_page)+1):
    page_url = '{}&page={}'.format(sise_url, page)  
    df = df.append(pd.read_html(page_url, header=0)[0])

df = df.dropna()
df = df.iloc[0:30] 
df = df.sort_values(by='날짜') 

plt.title('Celltrion (close)')
plt.xticks(rotation=45) 
plt.plot(df['날짜'], df['종가'], 'co-')
plt.grid(color='gray', linestyle='--')
plt.show()
```

```python
##전체 페이지 읽어오기(첫 페이지 ~ 마지막 페이지의 일별 시세 읽기)

#일별시세는 테이블 형테의 tabular데이터
df = pd.DataFrame() #일별 시세를 저장할 데이터프레임형 변수
sise_url = 'https://finance.naver.com/item/sise_day.nhn?code=005930' 

for page in range(1, int(last_page)+1): ##1~last page까지
    page_url = '{}&page={}'.format(sise_url, page)  #요청할 페이지 수 변경
    df = df.append(pd.read_html(page_url, header=0)[0])  #페이지 분량의 데이터프레임을 df 객체에 추가

df = df.dropna()##값이 빠진 행은 제거한다.
```

```python
# 차트 출력을 위한 데이터 프레임 가공
df = df.iloc[0:30] #모든 데이터 대신 최근 데이터 30행만 이용
df = df.sort_values(by='날짜') #날짜 오름차순 정렬

#차트 그리기
plt.title('Celltrion (close)')#제목
plt.xticks(rotation=45) #x축 레이블 회전(보기 편하게)
plt.plot(df['날짜'], df['종가'], 'co-')#x축 날짜 데이터 y출 종가 데이터, 좌표는 정로생 원으로
plt.grid(color='gray', linestyle='--')#각 좌표를 실선으로 연결해 표시
```

```python
##신버전으로 캔들 차트 그리기
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import mplfinance as mpf

url = 'https://finance.naver.com/item/sise_day.nhn?code=068270&page=1'
with urlopen(url) as doc:
    html = BeautifulSoup(doc, 'lxml') 
    pgrr = html.find('td', class_='pgRR')
    s = str(pgrr.a['href']).split('=')
    last_page = s[-1]  

df = pd.DataFrame()
sise_url = 'https://finance.naver.com/item/sise_day.nhn?code=068270'  
for page in range(1, int(last_page)+1): 
    page_url = '{}&page={}'.format(sise_url, page)  
    df = df.append(pd.read_html(page_url, header=0)[0])

df = df.dropna()
df = df.iloc[0:30]

###############################################################################################################
#한글 column명을 영어 column명으로 바꿈.
df = df.rename(columns={'날짜':'Date', '시가':'Open', '고가':'High', '저가':'Low', '종가':'Close', '거래량':'Volume'})
#날짜 오름차순 정렬
df = df.sort_values(by='Date')
#Data칼럼을 DateIndex형으로 변경한 것을 index로 설정
df.index = pd.to_datetime(df.Date)
#5개의 collumn만 갖도록 변경.
df = df[['Open', 'High', 'Low', 'Close', 'Volume']]
#캔들 차트 형태로 출력
mpf.plot(df, title='Celltrion candle chart', type='candle')
#OHLC 차트 형태로 출력(default)
mpf.plot(df, title='Celltrion ohlc chart', type='ohlc')

##캔들 색상 변경, 차트하단에 거래량 그래프 추가, 이동평균선 세개 지정
kwargs = dict(title='Celltrion customized chart', type='candle',
    mav=(2, 4, 6), volume=True, ylabel='ohlc candles')
mc = mpf.make_marketcolors(up='r', down='b', inherit=True) #상승은 빨간색, 하락은 파랑색으로 지정
s  = mpf.make_mpf_style(marketcolors=mc) #s: style객체
mpf.plot(df, **kwargs, style=s)
```
