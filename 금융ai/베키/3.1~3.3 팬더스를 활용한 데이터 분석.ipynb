## 3.1.1 배열 생성

- 넘파이 모듈은 리스트를 인수를 받아서 배열을 생성하는 `array()`함수를 제공
- 중첩된 리스트를 사용함

```python
# numpy import
import numpy as np

A = np.array([[1,2],[3,4]])
A
```

## 3.1.2 배열 정보 보기

- `type` : 배열의 자료형 확인
- `ndim` : 배열의 차원 확인
- `shape` : 배열의 크기(size) 확인
- `dtype` : 배열 속 원소의 자료형 확인

```python
# 배열의 자료형
print(type(A))
# 배열의 차원
print(A.ndim)
# 배열의 크기
print(A.shape)
# 원소 자료형
print(A.dtype)
```

```python
# 원소 별 최댓값, 평균값, 최솟값, 합계 구하는 함수도 제공
print(A.max(), A.mean(), A.min(), A.sum())
```

## 3.1.3 배열의 접근

- 리스트처럼 **인덱싱. 슬라이싱** 가능

- **A[행 인덱스][열 인덱스]** 도 가능하고 **A[행 인덱스, 열 인덱스]** 형식도 가능

```python
A = np.array([[1,2],[3,4]])

# A의 첫 번째 행
print(A[0])
print('--------'*5)

# A의 두 번째 열
print(A[:,1])  # 이건 열이지만
print(A[:][1]) # 이건 행이다. 
print(A[:]) 
print('--------'*5)

# A[행 인덱스][열 인덱스] == A[행 인덱스, 열 인덱스] 인 경우는 이렇게 위치가 정확히 지정될 때이다.
print(A[0][0],A[0][1])
print(A[1,0],A[1,1])
print('--------'*5)

# 조건에 맞는 원소만 인덱싱 가능
A[A>1]
```

## 3.1.4 배열 형태 바꾸기
1. **전치**(Transpose) : 배열의 원소 위치를 주대각선 기준으로 뒤바꾸기
    - `.T` 나 `transpose()` 함수를 사용해 배열의 전치를 구함.

```python
A = np.array([[1,2],[3,4]])

# .T 속성 사용
print(A.T)

# transpose()함수 사용
print(A.transpose())
```

2. **평탄화**(flatten) : 다차열 배열을 1차원 배열 형태로 바꾸기
    - `flatten()`함수를 사용해 배열을 1차원 배열로 바꾸기

```python
A = np.array([[1,2],[3,4]])

A.flatten()
```

## 3.1.5 배열의 연산

**같은 크기의 배열**끼리는 사칙 연산이 가능

```python
A = np.array([[1,2],
              [3,4]])

# np.add(A,A)와 같다.
print(A+A)
# np.substract(A,A)와 같다
print(A-A)
# np.multiply(A,A)와 같다.
print(A*A)
# np.divide(A,A)와 같다.
print(A/A)
```

## 3.1.6 브로드캐스팅

위처럼 크기가 같은 행렬에 대해서만 사칙 연산을 할 있다,  
넘파이에서는 행렬의 크기가 달라도 연산을 할 수 있게 **크기가 작은 행렬을 확장**해준다.   
이를 **브로드캐스팅(Broadcasting)**이라고 한다.

```python
A = np.array([[1,2],
              [3,4]])

B = np.array([10,100])

A*B
```

```python
C = A*B
C/A
```

## 3.1.7 내적 구하기

**내적(inner product)의 결과값 : 스칼라**

+ 넘파이의 배열끼리 곱셈(`*`) : 원소별로 곱셈 연산을 수행

+ 넘파이의 배열끼리 내적(`dot()`) : 배열의 내적 곱

(내적 보충 설명)
![내적](https://github.com/INVESTAR/StockAnalysisInPython/blob/master/03_NumPy_and_Pandas/imgs/InnerProduct.jpg?raw=true)

```python
B = np.array([10,100])

# 1차원 배열끼리 내적 곱
print(B.dot(B))

print(np.dot(B,B))   # 같은 결과
```

```python
A = np.array([[1,2],
              [3,4]])     #(2,2)행렬
B = np.array([10,100])    #(2,1)행렬

np.dot(A,B)    # 결과(2,1)행렬
```

## 3.2.1 Series 생성

시리즈에 인덱스를 별도로 지정하지 않으면, 0부터 시작하는 정수형 인덱스가 자동으로 생성된다.

```python
import pandas as pd

s = pd.Series([0, 3.6, 2, 5.8, 4.2, 8])
s
```

## 3.2.2 Series의 인덱스 변경

```python
# s 의 정수형 인덱스를 변경해보기
s.index = pd.Index([0,1.2,1.8,3,3.6,4.8])    #인덱스 변경
s.index.name = 'MY_IDX'                      #인덱스명 생성
s
```

```python
s.name = 'MY_SERIES'       #시리즈명 생성
s
```

## 3.2.3 데이터 추가
인덱스 레이블, 인덱스에 해당하는 값을 입력하면 추가됨.

```python
s[5.9] = 5.5
s
```

일일이 하나하나 추가하는게 힘들다면,  
**새로운 시리즈를 생성해서 append() 메서드로 추가**하는 방법도 있음,  
*(!주의! : 두 시리즈가 하나되면서 기존에 설정했던 시리즈명과 인덱스명이 초기화됨.)*

```python
ser = pd.Series([6.7,4.2], index=[6.8,8])    #ser 시리즈 생성
s = s.append(ser)    # 기존 s 에 신규 ser 시리즈를 추가
s
```

## 3.2.4 데이터 인덱싱

```python
# index 속성을 이용해 인덱스 값 구하기
s.index[-1]
```

```python
# values 속성을 이용해 인덱스 순서에 해당하는 데이터 구하기
s.values[-1]
```

- loc[인덱스값]   -> 로케이션 인덱서
- iloc[정수 순서] -> 인티저 로케이션 인덱서

```python
s.loc[8]
```

```python
s.iloc[-1]
```

**[주의] iloc과 values 모두 인덱스 순서에 해당하는 데이터를 출력한다는 점에서 동일하지만,  
반환하는  결과값이 복수 개 일때  
`iloc` -> 시리즈로 반환  
`values` -> 배열로 반환**

```python
s.iloc[:]  #시리즈로 반환
```

```python
s.values[:]  #배열로 반환
```

## 3.2.5 데이터 삭제

`drop()`메서드의 인수로 삭제

```python
s.drop(8)
```

```python
# 실제 s 에서는 위의 삭제가 반영되지 않음,
s
```

## 3.2.6 시리즈 정보 보기

`describe()` 메서드로 시리즈 객체의 대략적인 정보를 확인한다.

# 3.3 Pandas DataFrame

## 3.3.1 Dictionary를 이용한 DataFrame 새성

```python
import pandas as pd
df = pd.DataFrame({'KOSPI': [1915,1961,2026,2467,2041], 
                  'KOSDAQ': [542,682,631,798,675]},
                 index=[2014,2015,2016,2017,2018])
df
```

```python
df.describe()
```

```python
df.info()
```

## 3.3.2 Series를 이용한 DataFrame 생성

```python
# kospi 와 kosdaq 두 개의 시리즈를 각각 생성해서 데이터프레임으로 생성
kospi = pd.Series([1915,1961,2026,2467,2041],
                 index=[2014,2015,2016,2017,2018], name='KOSPI')
kospi
```

```python
kosdaq = pd.Series([542,682,631,798,675],
               index=[2014,2015,2016,2017,2018], name='KOSDAQ')
kosdaq
```

```python
# 생성된 두 개의 시리즈를 딕셔너리 형태로 구성해, 데이터프레임으로 만든다.
df = pd.DataFrame({kospi.name : kospi, kosdaq.name : kosdaq})
df
```

## 3.3.3 List를 이용한 DataFrame 생성

```python
# 리스트를 사용해 한 행씩 추가해서 데이터프레임을 생성한다.

columns = ['KOSPI','KOSDAQ']
index = [2014,2015,2016,2017,2018]
rows = []
rows.append([1915,542])
rows.append([1961,682])
rows.append([2026,631])
rows.append([2467,798])
rows.append([2041,675])
df = pd.DataFrame(rows, columns=columns, index=index)
df
```

## 3.3.4 데이터프레임 순회 처리

```python
for i in df.index:
    print(i, df['KOSPI'][i],df['KOSDAQ'][i])
```

**`itertuples()`메서드** : 데이터프레임의 각 행을 이름있는 튜플 형태로 반환(데이터프레임에서 제공하는 함수)

```python
for row in df.itertuples(name='KRX'):
    print(row)
```

```python
for row in df.itertuples(name='KRX'):
    print(row[0],row[1],row[2])
```

**`iterrows()`메서드** : 데이터프레임의 각 행을 인덱스와 시리즈 조합으로 반환(데이터프레임에서 제공하는 함수)

```python
for idx,row in df.iterrows():
    print(idx,row[0],row[1])
```
